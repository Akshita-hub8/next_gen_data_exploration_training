## 📖 What is Data?

Data is a collection of facts, observations, or measurements. It can be processed and analyzed to discover useful information and support decision-making.

### Why are Data Types Important?

Python needs to know the type of data before performing operations.

Common data types include:
- Integer (`int`)
- Float (`float`)
- String (`str`)
- Boolean (`bool`)

```python
# Integer
age = 21

# Float
cgpa = 8.7

# String
name = "Akshita"

# Boolean
is_placed = False

print("Age:", age, type(age))
print("CGPA:", cgpa, type(cgpa))
print("Name:", name, type(name))
print("Placed:", is_placed, type(is_placed))
```

## 📦 Python Data Structures

Data structures help us store and organize multiple values efficiently.

The four commonly used data structures are:
- **List** – Ordered and editable collection.
- **Tuple** – Ordered but cannot be modified.
- **Dictionary** – Stores data as key-value pairs.
- **Set** – Stores only unique values.

```python
# List Example
marks = [78, 85, 90, 67, 88]

print("Marks:", marks)
print("Highest:", max(marks))
print("Average:", sum(marks)/len(marks))
```

### 📝 Explanation

A **List** is used to store multiple values in a single variable.

In this example:
- `max()` returns the highest mark.
- `sum()` adds all marks.
- `len()` returns the total number of elements.
- `sum()/len()` calculates the average marks.

```python
# Tuple Example
months = ("Jan", "Feb", "Mar")

print("Months:", months)
print("Type:", type(months))
```

### 📝 Explanation

A **Tuple** is an ordered collection of items that cannot be modified after creation. It is useful for storing fixed data that should remain unchanged.

```python
# Dictionary Example
student = {
    "Name": "Akshita",
    "Age": 20,
    "CGPA": 8.7
}

print(student)
print("CGPA:", student["CGPA"])
```

### 📝 Explanation

A **Dictionary** stores data as **key-value pairs**.

It allows us to access values using their keys, making data retrieval fast and organized.

Example:
- **Key:** `"CGPA"`
- **Value:** `8.7`

```python
# Set Example
cities = {"Delhi", "Mumbai", "Delhi", "Pune"}

print(cities)
print(type(cities))
```

### 📝 Explanation

A **Set** stores only **unique** values and automatically removes duplicates.

It is useful when you need distinct items and don't want repeated values.

## 🐼 Introduction to Pandas

Pandas is a powerful Python library used for data analysis and manipulation.

It provides two main data structures:
- **Series** – One-dimensional data.
- **DataFrame** – Two-dimensional table (similar to an Excel sheet).

```python
import pandas as pd

# Pandas Series
ages = pd.Series([19, 20, 21, 22])

print(ages)
```

### 📝 Explanation

A **Series** is a one-dimensional data structure in Pandas. It stores a single column of data and is useful for working with sequences of values.

```python
# Pandas DataFrame
student_data = {
    "Name": ["Akshita", "Rahul", "Priya"],
    "Age": [20, 21, 19],
    "Marks": [91, 85, 88]
}

df = pd.DataFrame(student_data)

print(df)
```

### 📝 Explanation

A **DataFrame** is a two-dimensional table with rows and columns. It is one of the most commonly used data structures in Pandas for storing and analyzing datasets.

## 📊 Data Analytics Pipeline

A typical data analysis process follows these steps:

**Raw Data → Data Collection → Data Cleaning → Data Transformation → Data Analysis → Visualization → Insights**

This pipeline helps convert raw data into meaningful information for decision-making.

### 💡 Example

Suppose a college wants to analyze student performance.

- Raw Data → Student marks
- Data Cleaning → Remove missing values
- Analysis → Calculate average marks
- Visualization → Plot graphs
- Insights → Identify top-performing students

## 📈 Data Visualization

Data visualization helps present data in a graphical form, making it easier to identify patterns, trends, and relationships within a dataset.

### Common Visualization Techniques
- Bar Chart
- Histogram
- Box Plot
- Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

student_df = pd.DataFrame({
    "Name": ["Akshita", "Rahul", "Priya", "Aman", "Neha"],
    "CGPA": [8.7, 7.9, 9.1, 8.3, 8.8],
    "Department": ["CSE", "IT", "CSE", "ECE", "IT"]
})

student_df
```

### 📊 Bar Chart

A bar chart is used to compare values across different categories.

**Example:** Number of students in each department.

```python
plt.figure(figsize=(5,3))
sns.countplot(data=student_df, x="Department")

plt.title("Students by Department")
plt.show()
```

### 📈 Histogram

A histogram shows how numerical values are distributed.

**Example:** Distribution of students' CGPA.

```python
plt.figure(figsize=(5,3))
sns.histplot(student_df["CGPA"], bins=5)

plt.title("CGPA Distribution")
plt.show()
```

### 📦 Box Plot

A box plot helps detect outliers and understand the spread of numerical data.

**Example:** Checking the spread of CGPA values.

```python
plt.figure(figsize=(5,3))
sns.boxplot(y=student_df["CGPA"])

plt.title("CGPA Box Plot")
plt.show()
```

### ⭐ Observation

From these visualizations, we can quickly compare departments, understand the distribution of CGPA, and identify the spread of data using a box plot.
