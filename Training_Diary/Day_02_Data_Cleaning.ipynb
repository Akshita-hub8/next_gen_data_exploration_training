### 📖 What is Data Cleaning?

Data cleaning is the process of improving the quality of a dataset by identifying and fixing errors, inconsistencies, and incomplete information. It helps make the data accurate, reliable, and ready for analysis or machine learning.

Without proper cleaning, incorrect or missing information can lead to misleading results and poor decisions.

### ❓ Why is Data Cleaning Important?

Real-world datasets are rarely perfect. They often contain missing values, duplicate records, typing mistakes, or inconsistent formatting.

Data cleaning helps us:

- Improve the accuracy of analysis
- Reduce errors in reports
- Build reliable machine learning models
- Make better business decisions
- Save time during data analysis

Think of data cleaning as preparing ingredients before cooking. Clean ingredients lead to a better final dish, just as clean data leads to better insights.

### ⚠️ Common Data Quality Issues

While working with datasets, we may encounter several common problems:

- Missing Values
- Duplicate Records
- Incorrect Data Types
- Inconsistent Text Formatting
- Outliers
- Invalid Values
- Extra Spaces

These issues should be identified and corrected before performing any analysis.

### 📖 What are Missing Values?

Missing values are empty or unavailable entries in a dataset. They occur when some information is not recorded.

### Common Reasons
- Data not entered
- Survey questions skipped
- System or sensor errors

```python
students_df.isnull()
```

📝 **Explanation**

The `isnull()` function checks every cell in the dataset and returns **True** for missing values and **False** for available values.

```python
students_df.isnull().sum()
```

📝 **Explanation**

The `sum()` function counts the total number of missing values in each column.

```python
students_df[students_df.isnull().any(axis=1)]
```

📝 **Explanation**

This displays only those rows that contain one or more missing values.

### 🛠️ Methods to Handle Missing Values

Common ways to deal with missing values are:

- Remove Rows
- Fill with Mean
- Fill with Median
- Fill with Mode

```python
# Remove rows with missing values
students_df.dropna()
```

📝 **Explanation**

`dropna()` removes rows that contain missing values. It is useful when only a few records are incomplete.

```python
# Fill missing values with the mean
students_df["Age"] = students_df["Age"].fillna(students_df["Age"].mean())
students_df["CGPA"] = students_df["CGPA"].fillna(students_df["CGPA"].mean())
```

📝 **Explanation**

The **mean (average)** is used to fill missing values in numerical columns. It works well when the data is evenly distributed.

```python
# Fill missing values with the median
students_df["Attendance"] = students_df["Attendance"].fillna(students_df["Attendance"].median())
```

📝 **Explanation**

The **median** is the middle value of a dataset. It is a good choice when the data contains outliers.

```python
# Fill missing values with the mode
students_df["Gender"] = students_df["Gender"].fillna(students_df["Gender"].mode()[0])
students_df["City"] = students_df["City"].fillna(students_df["City"].mode()[0])
```

📝 **Explanation**

The **mode** is the most frequently occurring value. It is commonly used for categorical columns like Gender or City.

```python
# Verify that missing values have been handled
students_df.isnull().sum()
```

📝 **Explanation**

Run `isnull().sum()` again to confirm that the missing values have been filled or removed successfully.

### 📄 Duplicate Records

Duplicate records are rows that appear more than once in a dataset. They can affect calculations and reduce data quality.

```python
# Check duplicate rows
students_df.duplicated()
```

📝 **Explanation**

The `duplicated()` function returns **True** for duplicate rows and **False** for unique rows.

```python
# Count duplicate rows
students_df.duplicated().sum()
```

📝 **Explanation**

This counts the total number of duplicate rows in the dataset.

```python
# Remove duplicate rows
students_df = students_df.drop_duplicates()
```

📝 **Explanation**

`drop_duplicates()` removes repeated records and keeps only unique rows.

### ✍️ Cleaning Text Data

Text data may contain inconsistent capitalization or unwanted spaces. Cleaning text makes the data consistent and easier to analyze.

```python
# Standardize text formatting
students_df["Gender"] = students_df["Gender"].str.title()

# Remove extra spaces
students_df["Name"] = students_df["Name"].str.strip()

# Convert department names to uppercase
students_df["Department"] = students_df["Department"].str.upper()
```

📝 **Explanation**

- `title()` converts text to title case.
- `strip()` removes extra spaces.
- `upper()` converts text to uppercase for consistency.

### 📦 What are Outliers?

Outliers are values that are much higher or lower than the rest of the data. They can affect the accuracy of analysis and machine learning models.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,3))
sns.boxplot(y=students_df["CGPA"])

plt.title("CGPA Box Plot")
plt.show()
```

📝 **Explanation**

A **Box Plot** helps identify outliers by showing the spread of data. Points outside the box are possible outliers.

### 📊 Detecting Outliers Using IQR

The **Interquartile Range (IQR)** method is commonly used to detect outliers by calculating the spread between the first and third quartiles.

```python
Q1 = students_df["CGPA"].quantile(0.25)
Q3 = students_df["CGPA"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
outliers = students_df[
    (students_df["CGPA"] < lower_limit) |
    (students_df["CGPA"] > upper_limit)
]

outliers
```

📝 **Explanation**

This code finds all records where the CGPA is outside the acceptable IQR range.

```python
clean_df = students_df[
    (students_df["CGPA"] >= lower_limit) &
    (students_df["CGPA"] <= upper_limit)
]
```

```python
clean_df.info()
clean_df.describe()
```

📝 **Explanation**

After removing outliers, we check the dataset again to verify its structure and summary statistics.

### 💾 Saving the Cleaned Dataset

Once the data has been cleaned, it can be saved as a new CSV file for future analysis.

```python
clean_df.to_csv(
    "/content/drive/MyDrive/college_students_cleaned.csv",
    index=False
)
```

📝 **Explanation**

The `to_csv()` function saves the cleaned dataset as a new CSV file. Setting `index=False` prevents the row numbers from being saved.
